```python
alumnos = []

print("REGISTRO DE ALUMNOS")
print()

while True:
    nombre = input("Nombre: ")
    edad = int(input("Edad: "))
    nota = float(input("Nota: "))
    
    alumnos.append([nombre, edad, nota])
    
    continuar = input("¿Ingresar otro alumno? (s/n): ")
    print()
    if continuar == "n":
        break

print("\nLISTADO COMPLETO")
for a in alumnos:
    print(a[0], "-", a[1], "años - Nota:", a[2])

print("\nLISTADO ORDENADO POR NOTA")
for i in range(len(alumnos)):
    for j in range(i+1, len(alumnos)):
        if alumnos[i][2] < alumnos[j][2]:
            alumnos[i], alumnos[j] = alumnos[j], alumnos[i]

for a in alumnos:
    print(a[0], "- Nota:", a[2])

if len(alumnos) > 0:
    suma = 0
    for a in alumnos:
        suma = suma + a[2]
    print("\nPROMEDIO GENERAL")
    print("Promedio:", suma / len(alumnos))
```
